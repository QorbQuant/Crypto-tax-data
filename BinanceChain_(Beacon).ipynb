```python
import requests
import pandas as pd
```

```python
page = 1
rows = 25
address = "bnb18dykx9f6cy8t24zrfkhw0yre5rxcjtv90hy0js"
all_results = []

while True:
    url = f"https://explorer.bnbchain.org/api/v1/txs?page={page}&rows={rows}&address={address}"
    response = requests.get(url)
    if response.status_code != 200:
        print("Error:", response.text)
        break
    results = response.json()["txArray"]
    if not results:
        break
    all_results.extend(results)
    page += 1

df = pd.DataFrame(all_results)
print("Total results:", len(df))
```
